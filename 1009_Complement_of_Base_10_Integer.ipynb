### Solution in python

```python
class Solution:
    def bitwiseComplement(self, n: int) -> int:
        binary = format(n, 'b')
        # print((binary))

        st=""
        for i in binary:
            if i=='0':
                st = st+'1'
            else:
                st = st+'0'
        print(st)

        return int(st, 2)
```
