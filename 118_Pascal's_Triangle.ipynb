#### Solution 1 in python

```python
class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        l = [[1],[1,1]]
        c = []
        if numRows==1:
            return [[1]]
        elif numRows==2:
            return l
        else:
            for i in range(2, numRows):
                m = l[-1]
                c.append(1)
                for j in range(len(m)-1):
                    c.append(m[j]+m[j+1])
                c.append(1)
                l.append(c)
                c = []
            return l
```
