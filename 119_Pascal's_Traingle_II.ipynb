#### Solution 1 in python

```python
class Solution:
    def getRow(self, rowIndex: int) -> List[int]:
        l = [[1],[1,1]]
        c = []
        for i in range(rowIndex):
            m = l[-1]
            c.append(1)
            for j in range(len(m)-1):
                c.append(m[j]+m[j+1])
            c.append(1)
            l.append(c)
            c = []
        return l[rowIndex]
```
