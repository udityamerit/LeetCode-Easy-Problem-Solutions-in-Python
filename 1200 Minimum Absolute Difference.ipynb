#### Solution in python

```python
class Solution:
    def minimumAbsDifference(self, arr: List[int]) -> List[List[int]]:
        arr.sort()
        # print(arr)
        l = []
        c = min([abs(arr[i]-arr[i+1]) for i in range(len(arr)-1)])
        # print(c)
        return [[arr[i],arr[i+1]] for i in range(len(arr)-1) if abs(arr[i]-arr[i+1])==c]
        
```
