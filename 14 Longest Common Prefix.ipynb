#### Solution in python

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        pre = strs[0]
        for str in strs[1:]:
            while str[:len(pre)] != pre:
                # print(str[:len(pre)])
                pre = pre[:-1]
                # print(pre)

                if not pre:
                    return ""
        return pre
```
