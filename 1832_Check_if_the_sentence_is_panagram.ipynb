```python
from collections import Counter
class Solution:
    def checkIfPangram(self, sentence: str) -> bool:
        # l = [a, b, c, d, e, f, g, h, i, j, k, l, m, 
        # n, o, p, q, r, s, t, u, v, w, x, y, z]
        sen =list( set([i for i in sentence]))
        sen.sort()

        print(sen)
        if 26==len(sen):
            return True
```
