```python
class Solution:
    def canBeTypedWords(self, text: str, brokenLetters: str) -> int:
        words = text.split()
        broken_set = set(brokenLetters)
        count = 0
        
        for word in words:
            if all(char not in broken_set for char in word):
                count += 1

        return count
        
```

```python
class Solution:
    def canBeTypedWords(self, text: str, brokenLetters: str) -> int:
        words = text.split(' ')
        broken_set = set(brokenLetters)
        untypable_count = 0

        for word in words:
            for char in word:
                if char in broken_set:
                    untypable_count += 1
                    break 
        return len(words) - untypable_count
```
