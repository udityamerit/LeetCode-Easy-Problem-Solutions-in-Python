### Solution 1 in python

```python
class Solution:
    def maximumDifference(self, nums: List[int]) -> int:
        l = []
        for i in range(len(nums)):
            for j in range(i, len(nums)):
                if nums[i]<nums[j]:
                    l.append(nums[j]-nums[i])
        print(l)
        if len(l)==0:
            return -1
        else:
            return max(l)
        
            
```
