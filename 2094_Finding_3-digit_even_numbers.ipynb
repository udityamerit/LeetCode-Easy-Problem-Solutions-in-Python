#### Solution 1 in python


```python
class Solution:
    def findEvenNumbers(self, digits: List[int]) -> List[int]:
        result = set()
        for perm in itertools.permutations(digits, 3):
            number = perm[0]*100 + perm[1]*10 + perm[2]
            if number % 2 == 0 and perm[0] != 0:
                result.add(number)
        return sorted(result)
```
