#### Solution 1 in python

```python
class Solution:
    def divideString(self, s: str, k: int, fill: str) -> List[str]:
        p = []
        if len(s)%k==0:
           i = 0
           c = k
           while(i<len(s)):
            p.append(s[i:c])
            i = c
            c = c+k
        else:
            c = len(s)%k
            s = s+fill*(k-c)
            d = k
            i = 0
            while(i<len(s)):
                p.append(s[i:d])
                i = d
                d = d+k
        return p        
```
