```python
class Solution:
    def divideArray(self, nums: List[int]) -> bool:
        l = []
        nums.sort()
        # print(nums)
        
        if len(nums) % 2 != 0:
            return False 
        while nums:
            # print(nums[0], nums[1])
            if (nums[0] != nums[1]):
                return False
            l.append((nums[0], nums[1])) 
            nums.pop(0) 
            nums.pop(0)  

        # print(l)
        return True
```
