#### Solution 1 in Python

```python
class Solution:
    def maxFreqSum(self, s: str) -> int:
        mfr = Counter(s)
        v = 0
        c = 0
        ans = 0
        print(mfr)
        l = ['a', 'e', 'i', 'o', 'u']
        for key, value in mfr.items():
            if key in l:
                v = max(v, value)
            else:
                c = max(c,value)
            ans = v+c
        return ans
```
