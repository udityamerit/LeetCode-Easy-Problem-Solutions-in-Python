#### Solution 1 in python

```python
class Solution:
    def maximumNumberOfStringPairs(self, words: List[str]) -> int:
        c = 0
        for i in range(len(words)):
            for j in range(len(words)):
                if words[i]!=words[j] and words[i] == words[j][::-1]:
                    c = c+1
        return c//2
```
