#### Solution in python

```python
class Solution:
    def countSymmetricIntegers(self, low: int, high: int) -> int:
        c =0
        for i in range(low, high+1):
            if (len(str(i))%2==0):
                k = list(map(int, str(i)))
                # print(k)
                if(sum(k[:len(k)//2])==sum(k[len(k)//2:])):
                    c+=1
        return c

        
```
