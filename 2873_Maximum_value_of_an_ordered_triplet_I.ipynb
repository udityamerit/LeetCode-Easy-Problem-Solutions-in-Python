#### Solution in Python

```python
class Solution:
    def maximumTripletValue(self, nums: List[int]) -> int:
        l = []
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                for k in range(j+1, len(nums)):
                    l.append((nums[i] - nums[j]) * nums[k])
        print(l)
        if max(l)>0:
            return max(l)
        else:
            return 0
        
```
