#### Solution 1 in python

```python
class Solution:
    def countSubarrays(self, nums: List[int]) -> int:
        l = 0
        n = len(nums)
        for k in range(3,4):
            for i in range(n-k+1):
                subarr = nums[i:i+k]
                # print(subarr)
                if subarr[0]+subarr[2] == subarr[1]/2:
                    # print(subarr[1]/2)
                    l = l+1
        # print(l)
        return l
        
```
