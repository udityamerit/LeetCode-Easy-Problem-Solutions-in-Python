#### Solution 1 in Python


```python
class Solution:
    def maxAdjacentDistance(self, nums: List[int]) -> int:
        max_diff = -inf
        for i in range(len(nums)-1):
                a = abs(nums[i]-nums[i+1])
                # print(a)
                if a>max_diff:
                    max_diff=a
        
        if max_diff>abs(nums[0]-nums[len(nums)-1]):
            return max_diff
        else:
            return abs(nums[0]-nums[len(nums)-1])
```

```python
class Solution:
    def maxAdjacentDistance(self, nums: List[int]) -> int:
        #### Solution 1
        nums.append(nums[0])
        n = len(nums)
        return max([abs(nums[i]-nums[i+1]) for i in range(n-1)])
```
