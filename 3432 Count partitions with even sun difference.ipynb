#### Solution in python

```python
class Solution:
    def countPartitions(self, nums: List[int]) -> int:
        count  =0
        n = len(nums)
        # print(sum(nums[0]))
        for i in range(1,len(nums)):
            if ((sum(nums[: i])-sum(nums[i:]))%2==0):
                # print(((nums[0: i]), (nums[i: n])))
                count+=1
            # if count==0:
            #     return 0
        return count
```
