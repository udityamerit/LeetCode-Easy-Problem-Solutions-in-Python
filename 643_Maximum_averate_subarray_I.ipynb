```python
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        n = len(nums)
        l = -inf
        # for i in range(n-k+1):
        #     m=sum(nums[i:i+k])/k
        #     # print(m)
        #     if m>l:
        #         l = m
        # return l

        current_sum = sum(nums[:k])
        max_sum = current_sum
        
        for i in range(k, len(nums)):
            current_sum += nums[i] - nums[i - k]
            max_sum = max(max_sum, current_sum)
        
        return max_sum / k
```
