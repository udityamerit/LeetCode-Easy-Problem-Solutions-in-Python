#### Solution 1

```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:     
        for i in range(len(matrix)):
            if target in matrix[i]:
                return True 
```

#### Solution 2 in python


```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        for i in range(len(matrix)):
            for j in range(len(matrix[i])):
                if matrix[i][j]==target:
                    return True
                else:
                    return False
```
