#### Solution 1 in python

```python
class Solution:
    def rotateString(self, s: str, goal: str) -> bool:
        if len(s) != len(goal):
            return False
        if s == goal:
            return True
            
        t = list(s)
        for i in range(len(s)):
            t.append(t.pop(0)) 
            
            if "".join(t) == goal:
                return True
        
        
        return False
```
