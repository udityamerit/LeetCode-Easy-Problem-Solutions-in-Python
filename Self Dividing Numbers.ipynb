## Solution 1 in python

```python
class Solution:
    def selfDividingNumbers(self, left: int, right: int) -> List[int]:
        l = []
        for i in range(left, right+1):
            c = len(str(i))
            if '0' not in str(i):
                for j in str(i):
                    if j!=0 and i%int(j)==0:
                        c = c-1
                    
                if c==0:
                    l.append(i)
           
        return l
```
